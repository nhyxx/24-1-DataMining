# 데이터 불러오기 및 데이터 전처리

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
    
kream = pd.read_csv('df_filtered.csv', encoding = 'cp949')
kream
```

```python
kream_2 = kream[['수익률', '거래량', '발매일로부터현재까지일']]  #필요한 데이터 분리
print(kream_2)
```

```python
kream_2 = kream_2.rename(columns = {'수익률': 'rate', '거래량': 'volume', '발매일로부터현재까지일': 'duration'})    #열이름변경
print(kream_2)
```

```python
print("max\n", kream_2.max())
print("min\n", kream_2.min())
```

```python
from sklearn.preprocessing import StandardScaler # 스케일링

scaler = StandardScaler()
kream_2_scaled = scaler.fit_transform(kream_2)
print(kream_2_scaled)
```

## 조건이 '수익률', '거래량', '발매일부터 현재까지의 기간' 일 때
### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

### 그래프 그리기

```python
import matplotlib.font_manager as fm

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] =False
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = '3d')

x = kream_2['rate']
y = kream_2['volume']
z = kream_2['duration']

ax.scatter(x, y, z, c = 'g', edgecolors = 'red')
ax.set_xlabel('수익률')
ax.set_ylabel('거래량')
ax.set_zlabel('가격변동성')

plt.show()
```

### DBSCAN clustering

```python
eps_settings = [0.1, 0.25, 0.5, 1, 2]
ms_settings = [2, 3, 5, 7]

for elps in eps_settings:
    for ms in ms_settings:
        db = DBSCAN(eps = elps, min_samples = ms)
        db.fit(kream_2_scaled)       
                
        kream_2_db = db.labels_
        
        print("eps = ", elps, "min_samples = ", ms)
        print(pd.DataFrame(kream_2_db).value_counts(sort = False))
```

#### eps =  0.25, min_samples =  7 일 때, 적당하다고 판단 

```python
dbscan = DBSCAN(eps =  0.25, min_samples =  7)
dbscan.fit(kream_2_scaled)
```

```python
kream_2_dbscan = dbscan.labels_
print(pd.DataFrame(kream_2_dbscan))
```

```python
kream_2_dbscan = pd.DataFrame(kream_2_dbscan)
kream_2_dbscan.value_counts()  #클러스터별 개수 파악
```

```python
kream_2.insert(3, 'dbscan', kream_2_dbscan)
kream_2_clustered = kream_2  #클러스터라벨링
```

```python
kream_2_clustered
```

### 그래프 그리기

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = '3d')

x = kream_2_clustered['rate']
y = kream_2_clustered['volume']
z = kream_2_clustered['duration']

scatter = ax.scatter(x, y, z, c = kream_2_clustered.dbscan, cmap = "rainbow", alpha = 0.5)
ax.set_xlabel('수익률')
ax.set_ylabel('거래량')
ax.set_zlabel('발매일로부터 현재까지의 기간')

legend = ax.legend(*scatter.legend_elements(), loc="upper right", title="Clusters")
ax.add_artist(legend)
plt.show()
```

```python
scatter = plt.scatter(kream_2_clustered['rate'], kream_2_clustered['volume'], c = kream_2_clustered['dbscan'], cmap = "rainbow", alpha = 0.5)
plt.xlabel('수익률')
plt.ylabel('거래량')
plt.legend(*scatter.legend_elements(), title="Clusters")
```

```python
scatter = plt.scatter(kream_2_clustered['volume'], kream_2_clustered['duration'], c = kream_2_clustered['dbscan'], cmap = "rainbow", alpha = 0.5)
plt.xlabel('거래량')
plt.ylabel('발매일로부터 현재까지의 기간')
plt.legend(*scatter.legend_elements(), title="Clusters")
```

### 결과와 관련된 통계량

```python
print("평균값\n", kream_2_clustered[['rate', 'volume', 'duration', 'dbscan']].groupby('dbscan').mean())
print("\n표준편차\n", kream_2_clustered[['rate', 'volume', 'duration', 'dbscan']].groupby('dbscan').std())
print("\n중앙값\n", kream_2_clustered[['rate', 'volume', 'duration', 'dbscan']].groupby('dbscan').median())
print("\n최대값\n", kream_2_clustered[['rate', 'volume', 'duration', 'dbscan']].groupby('dbscan').max())
print("\n최소값\n", kream_2_clustered[['rate', 'volume', 'duration', 'dbscan']].groupby('dbscan').min())
```
