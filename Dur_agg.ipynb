# 데이터 불러오기 및 데이터 전처리

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
    
kream = pd.read_csv('df_filtered.csv', encoding = 'cp949')
kream
```

```python
kream_2 = kream[['수익률', '거래량', '발매일로부터현재까지일']]  #필요한 데이터 분리
print(kream_2)
```

```python
kream_2 = kream_2.rename(columns = {'수익률': 'rate', '거래량': 'volume', '발매일로부터현재까지일': 'duration'})    #열이름변경
print(kream_2)
```

```python
print("max\n", kream_2.max())
print("min\n", kream_2.min())
```

```python
from sklearn.preprocessing import StandardScaler # 스케일링

scaler = StandardScaler()
kream_2_scaled = scaler.fit_transform(kream_2)
print(kream_2_scaled)
```

## 조건이 '수익률', '거래량', '발매일부터 현재까지의 기간' 일 때
### AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering
```

### 그래프 그리기

```python
import matplotlib.font_manager as fm

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] =False
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = '3d')

x = kream_2['rate']
y = kream_2['volume']
z = kream_2['duration']

ax.scatter(x, y, z, c = 'g', edgecolors = 'red')
ax.set_xlabel('수익률')
ax.set_ylabel('거래량')
ax.set_zlabel('가격변동성')

plt.show()
```

### AgglomerativeClustering

```python
nc_settings = [2, 3, 4, 5]
linkage_settings = ['ward', 'complete', 'average', 'single']

for nc in nc_settings:
    for link in linkage_settings:
        aggl = AgglomerativeClustering(n_clusters = nc, linkage = link)
        aggl.fit(kream_2_scaled)

        kream_2_agg = aggl.labels_
        print("n_clusters = ", nc, "linkage = ", link)
        print(pd.DataFrame(kream_2_agg).value_counts(sort = False))
```

#### n_clusters =  4, linkage =  'ward' 일 때, 적당하다고 판단 

```python
agg = AgglomerativeClustering(n_clusters =  4, linkage =  'ward')
agg.fit(kream_2_scaled)
```

```python
kream_2_agg = agg.labels_
print(pd.DataFrame(kream_2_agg))
```

```python
kream_2_agg = pd.DataFrame(kream_2_agg)
kream_2_agg.value_counts() # 클러스터별 개수 확인
```

```python
kream_2.insert(3, 'agg', kream_2_agg) #클러스터 라벨링
kream_2_clustered = kream_2
```

```python
kream_2_clustered
```

### 그래프 그리기

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = '3d')

x = kream_2_clustered['rate']
y = kream_2_clustered['volume']
z = kream_2_clustered['duration']

scatter = ax.scatter(x, y, z, c = kream_2_clustered['agg'], cmap = "rainbow", alpha = 0.5)
ax.set_xlabel('수익률')
ax.set_ylabel('거래량')
ax.set_zlabel('발매일로부터 현재까지의 기간')

legend = ax.legend(*scatter.legend_elements(), loc="upper right", title="Clusters")
ax.add_artist(legend)
plt.show()
```

```python
scatter = plt.scatter(kream_2_clustered['rate'], kream_2_clustered['volume'], c = kream_2_clustered['agg'], cmap = "rainbow", alpha = 0.5)
plt.xlabel('수익률')
plt.ylabel('거래량')
plt.legend(*scatter.legend_elements(), title="Clusters")
```

### 결과와 관련된 통계량

```python
print("평균값\n", kream_2_clustered[['rate', 'volume', 'duration', 'agg']].groupby('agg').mean())
print("\n표준편차\n", kream_2_clustered[['rate', 'volume', 'duration', 'agg']].groupby('agg').std())
print("\n중앙값\n", kream_2_clustered[['rate', 'volume', 'duration', 'agg']].groupby('agg').median())
print("\n최대값\n", kream_2_clustered[['rate', 'volume', 'duration', 'agg']].groupby('agg').max())
print("\n최소값\n", kream_2_clustered[['rate', 'volume', 'duration', 'agg']].groupby('agg').min())
```
