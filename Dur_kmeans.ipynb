# 데이터 불러오기 및 데이터 전처리

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
    
kream = pd.read_csv('df_filtered.csv', encoding = 'cp949')
kream
```

```python
kream_2 = kream[['수익률', '거래량', '발매일로부터현재까지일']]  #필요한 데이터 분리
print(kream_2)
```

```python
kream_2 = kream_2.rename(columns = {'수익률': 'rate', '거래량': 'volume', '발매일로부터현재까지일': 'duration'})    #열이름변경
print(kream_2)
```

```python
print("max\n", kream_2.max())
print("min\n", kream_2.min())
```

```python
from sklearn.preprocessing import StandardScaler #스케일링

scaler = StandardScaler()
kream_2_scaled = scaler.fit_transform(kream_2)
print(kream_2_scaled)
```

## 조건이 '수익률', '거래량', '발매일부터 현재까지의 기간' 일 때
### KMeans

```python
from sklearn.cluster import KMeans
```

### 그래프 그리기

```python
import matplotlib.font_manager as fm

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] =False
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = '3d')

x = kream_2['rate']
y = kream_2['volume']
z = kream_2['duration']

ax.scatter(x, y, z, c = 'g', edgecolors = 'red')
ax.set_xlabel('수익률')
ax.set_ylabel('거래량')
ax.set_zlabel('가격변동성')

plt.show()
```

### K-means clustering 적용

elbow method를 이용하여 k값 찾기

```python
from sklearn.metrics import silhouette_score

inertia = []
sil_scores = []

for n_clusters in range(2,15):
    kmeans = KMeans(n_clusters = n_clusters, random_state = 10)
    cluster_labels = kmeans.fit_predict(kream_2_scaled)


    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(kream_2_scaled, cluster_labels))
```

```python
plt.plot(range(2,15), inertia, marker='o', lw=2, ms=10)
plt.xlabel("Number of clusters")
plt.title("K-Means Inertia")
```

```python
sil_scores # 실루엣 스코어
```

```python
plt.bar(range(2,15), sil_scores)
plt.xlabel("Number of clusters")
plt.title("Silhouette Score")
```

inertia와 silhouette score을 고려해봤을 때 높지는 않지만 k = 2이 적절하다고 판단.

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(kream_2_scaled)

print(kmeans.cluster_centers_)

kmeans_labels = kmeans.labels_
print(kmeans_labels)
```

```python
kream_2_kmeans = pd.DataFrame(kmeans_labels)
kream_2_kmeans.value_counts() # 클러스터별 개수
```

```python
kream_2.insert(3, 'kmeans', kream_2_kmeans)
kream_2_clustered = kream_2  #클러스터 라벨링
```

```python
kream_2_clustered
```

### 그래프 그리기

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = '3d')

x = kream_2_clustered['rate']
y = kream_2_clustered['volume']
z = kream_2_clustered['duration']

scatter = ax.scatter(x, y, z, c = kream_2_clustered.kmeans, cmap = "rainbow", alpha = 0.5)
ax.set_xlabel('수익률')
ax.set_ylabel('거래량')
ax.set_zlabel('발매일로부터 현재까지의 기간')

legend = ax.legend(*scatter.legend_elements(), loc="upper right", title="Clusters")
ax.add_artist(legend)
plt.show()
```
